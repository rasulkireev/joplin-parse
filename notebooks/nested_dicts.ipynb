```python
def has_children(folder):
    try:
        folder['children']
        return True
    except KeyError:
        return False
```

```python
import re

def search_for_joplin_links(text):
    internal_links = re.findall("\(:/.+\)", text)

    link_ids = []
    for link in internal_links:
        note_id = re.search("(?<=\(:\/).+?(?=\))", link).group()
        link_ids.append(note_id)
    
    return link_ids

def find_linked_note(notes, link):
    for note in notes:
        if note['id'] == link:
            linked_note = note
    
    return linked_note
```

```python
def search_for_note_parent_folder(folders, notes_parent_id):
    for folder in folders:
        if folder['id'] == notes_parent_id['parent_id']:
            searched_folder = folder
        elif not folder['id'] == notes_parent_id['parent_id']:
            if has_children(folder):
                folders = folder['children'] 
                for folder in folders:
                    if folder['id'] == notes_parent_id['parent_id']:
                        searched_folder = folder
                    elif not folder['id'] == notes_parent_id['parent_id']:
                        if has_children(folder):
                            folders = folder['children'] 
                            for folder in folders:
                                if not folder['id'] == notes_parent_id['parent_id']:
                                    continue
                                elif folder['id'] == notes_parent_id['parent_id']:
                                    searched_folder = folder
    return searched_folder
                                    
                                    
def search_for_folder_parent_folder(folders, folder_parent_id):
    for folder in folders:
        if folder['id'] == folder_parent_id['parent_id']:
            searched_folder = folder
        elif not folder['id'] == folder_parent_id['parent_id']:
            if has_children(folder):
                folders = folder['children'] 
                for folder in folders:
                    if folder['id'] == folder_parent_id['parent_id']:
                        searched_folder = folder
                    elif not folder['id'] == folder_parent_id['parent_id']:
                        if has_children(folder):
                            folders = folder['children'] 
                            for folder in folders:
                                if not folder['id'] == folder_parent_id['parent_id']:
                                    continue
                                elif folder['id'] == folder_parent_id['parent_id']:
                                    searched_folder = folder
                
    return searched_folder
                 
```

```python
# search for note with title "The Five Most Common Writing Mistakes - Stew Fortier"

for note in notes:
    if "The Five Most Common Writing Mistakes" in note['title']:
        break
    else:
        continue
        
link_ids = search_for_joplin_links(note['body'])

for link in link_ids:
    linked_note = find_linked_note(notes, link)
    print(linked_note)
```
